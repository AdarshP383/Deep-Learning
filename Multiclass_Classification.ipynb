# Multiclass classification
Number classes is more than 2 then it is a multi class clasification

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd
url = ('https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/main/mobile_train.csv')
df = pd.read_csv(url)
df.head()
```

```python
df ['price_range'].unique()
```

# Price range

0 - Cheap
1 - Moderate
2 - High
3 - Very Expensive


### 4 classes to predict in above dataframe


### Step 3: Seperate X and Y(price_range)

```python
x = df.drop(columns=['price_range'])
y = df[['price_range']]
```

```python
x.head()
```

```python
y.head()
```

### Step 4: Perform preprocessing on X

```python
x.dtypes
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
x_pre =num_pipe.fit_transform(x)
```

```python
x_pre.head()
```

```python
x.shape
```

### Step 5: apply train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x_pre, y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### This is applicable to multiclass classification only, get onehot encoding for target features

```python
from keras.utils import to_categorical
ytrain_cat = to_categorical(ytrain)
ytest_cat = to_categorical(ytest)
```

```python
ytrain.head()
```

```python
ytrain_cat[0:5]
```

```python
ytest_cat [0:5]
```

```python
ytrain_cat.shape[1]
```

### Step 6: Build the neural network

image.png

```python
ytrain_cat.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
from keras.callbacks import EarlyStopping
```

```python
# Initilaze the model
model = Sequential()

# Create a Input layer
model.add(Input (shape=(xtrain.shape[1],)))

#Create Dense layer 1
model.add (Dense (64, activation='relu'))
#Add dropout layer
model.add(Dropout(0.2))

#Create Dense layer 2
model.add (Dense (32, activation='relu'))
# Add dropout layer
model.add(Dropout(0.1))

#Output layer 3
model.add (Dense (ytrain_cat.shape[1], activation='softmax'))
```

```python
model.summary()
```

```python
model.compile (optimizer='adam',
               loss = 'categorical_crossentropy',
               metrics=['accuracy'])
```

```python
# Add early stopping

es = EarlyStopping(monitor='val_loss', patience=15)
```

```python
hist = model.fit (
    xtrain,
    ytrain_cat,
    validation_split=0.2,
    epochs=200,
    callbacks=[es]
)
```

```python
import matplotlib.pyplot as plt
plt.figure (figsize=(12,6))
plt.plot(hist.history['loss'], label = "Train Loss")
plt.plot (hist.history ['val_loss'], label = 'Validation Loss')
plt.xlabel('Ephocs')
plt.ylabel ('Categorical Crossentropy')
plt. title ('Learning Curve ')
plt.show()
```

### Model Evalatuation

```python
model.evaluate (xtest, ytest_cat)
```

```python
model.evaluate (xtrain, ytrain_cat)
```

```python
### Predict the results for train and test
```

```python
yprob_train = model.predict(xtrain)
yprob_test = model.predict(xtest)
```
